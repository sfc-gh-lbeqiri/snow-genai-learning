## 1.1 Gen AI & LLM Concepts

**Flashcard 1** — Token

| Key | Detail |
|---|---|
| **Concept** | Token |
| **Detail** | The smallest unit of text an LLM processes. A token is roughly 4 characters or 0.75 words. Use `AI_COUNT_TOKENS` to estimate before calling an LLM. |
| **Lab** | 01–02 |

**Flashcard 2** — Embedding

| Key | Detail |
|---|---|
| **Concept** | Embedding |
| **Detail** | A dense numerical vector that represents the semantic meaning of text or images. Created with `AI_EMBED`. Similar meanings produce vectors with high cosine similarity. |
| **Lab** | 01–02 |

**Flashcard 3** — Temperature

| Key | Detail |
|---|---|
| **Concept** | Temperature |
| **Detail** | Controls randomness in LLM output (0.0–1.0). 0 = deterministic, 1 = creative. Set via `model_parameters` in `AI_COMPLETE`. |
| **Lab** | 01–02 |

**Flashcard 4** — top_p (Nucleus Sampling)

| Key | Detail |
|---|---|
| **Concept** | top_p (Nucleus Sampling) |
| **Detail** | Alternative to temperature. Restricts output to the smallest set of tokens whose cumulative probability exceeds p. Lower values = more focused output. |
| **Lab** | 01–02 |

**Flashcard 5** — max_tokens

| Key | Detail |
|---|---|
| **Concept** | max_tokens |
| **Detail** | Limits the maximum number of tokens in the response. Default 4096, max 8192. Prevents runaway costs on long completions. |
| **Lab** | 01–02 |

**Flashcard 6** — Context Window

| Key | Detail |
|---|---|
| **Concept** | Context Window |
| **Detail** | The maximum combined token count for prompt + response that a model can handle. Exceeding it causes truncation or errors. |
| **Lab** | 01–02 |

## 1.2 Prompt Engineering Patterns

**Flashcard 7** — Zero-Shot

| Key | Detail |
|---|---|
| **Pattern** | Zero-Shot |
| **Description** | Ask the model directly without examples. Works for general tasks. |
| **Example** | `AI_COMPLETE('model', 'Classify: positive or negative: I love this product')` |
| **Lab** | 01 |

**Flashcard 8** — Few-Shot

| Key | Detail |
|---|---|
| **Pattern** | Few-Shot |
| **Description** | Provide 2–3 examples in the prompt so the model learns the pattern. |
| **Example** | `Prompt includes: 'Example 1: input → output. Example 2: input → output. Now classify: ...'` |
| **Lab** | 01 |

**Flashcard 9** — Chain-of-Thought

| Key | Detail |
|---|---|
| **Pattern** | Chain-of-Thought |
| **Description** | Ask the model to reason step by step before answering. Improves accuracy on complex tasks. |
| **Example** | `'Think step by step. What is the total cost if...'` |
| **Lab** | 01 |

**Flashcard 10** — System Context

| Key | Detail |
|---|---|
| **Pattern** | System Context |
| **Description** | Embed persona and constraints at the start of the prompt. With AI_COMPLETE, this goes in the prompt string directly. |
| **Example** | `'You are a Snowflake expert. Answer in bullet points. ...'` |
| **Lab** | 01 |

## Practice Questions

**Q1.** What is a token in the context of LLMs?

A) A security credential for API access
B) The smallest unit of text an LLM processes, roughly 4 characters
C) A unique identifier for each model
D) A billing unit equal to one API call

> **Answer: B** — Tokens are sub-word text units. AI_COUNT_TOKENS estimates them. One token ≈ 4 chars.

**Q2.** Which parameter controls randomness in AI_COMPLETE output?

A) max_tokens
B) guardrails
C) temperature
D) response_format

> **Answer: C** — temperature (0–1) controls randomness. 0 = deterministic, higher = more creative. top_p is the alternative.

**Q3.** What does AI_EMBED return?

A) A STRING summary
B) A BOOLEAN
C) A VECTOR
D) An OBJECT with labels

> **Answer: C** — AI_EMBED returns a VECTOR. Dimension depends on model (768 or 1024).

**Q4.** In the RAG pattern, what is the correct order of steps?

A) Generate → Search → Embed
B) Search → Embed → Generate
C) Embed → Search → Generate
D) Embed → Generate → Search

> **Answer: C** — RAG = Embed the query → Search for similar context → Generate a grounded answer with AI_COMPLETE.

**Q5.** What is the primary advantage of RAG over fine-tuning?

A) Lower latency
B) No training data needed and knowledge can be updated without retraining
C) Better grammar
D) Works without a warehouse

> **Answer: B** — RAG retrieves from a live knowledge base — no model retraining. Fine-tuning requires training data and compute.

**Q6.** Which function estimates token count before calling an LLM?

A) AI_EMBED
B) AI_COMPLETE with show_details
C) AI_COUNT_TOKENS
D) AI_FILTER

> **Answer: C** — AI_COUNT_TOKENS('ai_complete', 'model', 'text') returns the estimated token count.

**Q7.** What is the first argument to AI_COUNT_TOKENS?

A) The model name
B) The input text
C) A lowercase function name string like 'ai_complete'
D) The max token limit

> **Answer: C** — AI_COUNT_TOKENS requires the function name (e.g. 'ai_complete') as the first arg, then model, then text.

**Q8.** What does 'zero-shot' prompting mean?

A) Using zero tokens
B) Providing no examples — just the task instruction
C) Setting temperature to 0
D) Calling the model with no prompt

> **Answer: B** — Zero-shot = direct task instruction with no examples. Few-shot = providing examples in the prompt.

**Q9.** Which prompt engineering pattern asks the model to 'think step by step'?

A) Zero-shot
B) Few-shot
C) Chain-of-thought
D) Extraction

> **Answer: C** — Chain-of-thought prompting improves reasoning by asking the model to show its work step by step.

**Q10.** What does Cortex Search Service manage automatically?

A) Model training
B) Embedding, indexing, and retrieval
C) RBAC grants
D) Cost budgets

> **Answer: B** — Cortex Search handles embedding + indexing + retrieval via a single CREATE CORTEX SEARCH SERVICE DDL.

**Q11.** What does TARGET_LAG control in a Cortex Search Service?

A) Query timeout
B) How often the index refreshes from source data
C) Maximum result count
D) Network latency threshold

> **Answer: B** — TARGET_LAG sets the maximum staleness of the search index (e.g. '1 minute').

**Q12.** When should you choose fine-tuning over prompting?

A) When you need real-time data access
B) When you need consistent domain-specific tone and format that prompting can't achieve
C) When you want the cheapest option
D) When the model is too slow

> **Answer: B** — Fine-tuning bakes domain behavior into the model. Prompting is cheaper but less consistent for specialized tasks.

**Q13.** What does FINETUNE('CREATE', ...) do?

A) Creates a new base model from scratch
B) Starts a fine-tuning job that produces a custom model from training data
C) Creates a Cortex Search index
D) Enables guardrails on a model

> **Answer: B** — FINETUNE creates a training job. It requires a training data table with prompt/completion pairs.

**Q14.** Which Snowflake feature orchestrates Search + Analyst + AI functions together?

A) Dynamic Tables
B) Cortex Agent
C) Cortex Guard
D) Streams

> **Answer: B** — Cortex Agent routes user questions to the right tool (Search, Analyst, AI functions) via REST API.

**Q15.** What is the key difference between an embedding model and a language model?

A) Embedding models generate text; language models generate vectors
B) Embedding models convert input to dense vectors for similarity; language models generate text completions
C) They are the same thing
D) Language models are only used for translation

> **Answer: B** — Embedding models (AI_EMBED) create numerical vector representations for similarity search. Language models (AI_COMPLETE) generate text responses from prompts.

**Q16.** How do LLM models run in Snowflake Cortex?

A) Data is sent to external APIs
B) Models run on customer's local machines
C) Fully hosted within Snowflake — data never leaves the platform
D) Models run in a third-party cloud

> **Answer: C** — All Cortex models run within Snowflake's infrastructure. Data stays in-platform for security and governance.

**Q17.** What is cross-region inference?

A) Querying data across databases
B) Running a model in a different Snowflake region when it's not available locally
C) Replicating models to edge locations
D) Load balancing across warehouses

> **Answer: B** — CORTEX_ENABLED_CROSS_REGION allows models to execute in another region if unavailable in the account's region.

**Q18.** What is the difference between AI_COMPLETE's simple form and show_details form?

A) Simple returns OBJECT, show_details returns STRING
B) Simple returns STRING, show_details returns OBJECT with choices + usage metadata
C) They return the same thing
D) show_details only works with image inputs

> **Answer: B** — AI_COMPLETE(model, prompt) → STRING. With show_details => true → OBJECT with choices[0]:messages + usage tokens.

**Q19.** How many vector distance functions does Snowflake provide?

A) 1
B) 2
C) 3
D) 4

> **Answer: D** — Cosine similarity, inner product, L1 (Manhattan), and L2 (Euclidean) distance.

**Q20.** What is the role of embeddings in semantic search?

A) They encrypt the data
B) They convert text to numerical vectors so similarity can be measured mathematically
C) They compress data for storage
D) They translate text between languages

> **Answer: B** — Embeddings map text to dense vectors. Similar texts have close vectors (high cosine similarity).

**Q21.** Which AI_EMBED model produces 1024-dimensional vectors?

A) e5-base-v2
B) snowflake-arctic-embed-m
C) snowflake-arctic-embed-l-v2.0
D) mistral-large2

> **Answer: C** — snowflake-arctic-embed-l-v2.0 produces 1024-dim vectors. e5-base-v2 and snowflake-arctic-embed-m produce 768-dim.

**Q22.** What is the max_tokens default for AI_COMPLETE?

A) 1024
B) 2048
C) 4096
D) 8192

> **Answer: C** — AI_COMPLETE is the preferred replacement for SNOWFLAKE.CORTEX.COMPLETE. It supports text, image, and prompt object inputs. The old function still works but is considered legacy.

**Q23.** Which function replaced SNOWFLAKE.CORTEX.COMPLETE?

A) AI_FILTER
B) AI_AGG
C) AI_COMPLETE
D) AI_SUMMARIZE_AGG

> **Answer: C** — AI_COMPLETE is the preferred replacement. The old SNOWFLAKE.CORTEX.COMPLETE still works but is legacy.

**Q24.** What does a Verified Query (VQR) do in Cortex Analyst?

A) Validates SQL syntax
B) Provides a known-correct SQL query for a natural language question
C) Runs a security scan
D) Counts tokens

> **Answer: B** — VQRs are curated question→SQL pairs that improve Analyst accuracy for common questions.

**Q25.** What is the maximum image size supported by AI_COMPLETE for vision tasks?

A) 1 MB
B) 5 MB
C) 10 MB (3.75 MB for Claude)
D) 50 MB

> **Answer: C** — Max 10 MB for most models, 3.75 MB for Claude. Supported formats: JPEG, PNG, WEBP, GIF, BMP.
