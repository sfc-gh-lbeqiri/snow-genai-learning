## Practice Questions

**Q1.** What does AI_COMPLETE(model, prompt) return?

A) OBJECT
B) BOOLEAN
C) STRING
D) VECTOR

> **Answer: C** — The simple form returns a STRING directly. Use show_details => true to get an OBJECT.

**Q2.** How do you get token usage metadata from AI_COMPLETE?

A) AI_COUNT_TOKENS
B) show_details => true
C) response_format => 'json'
D) guardrails => true

> **Answer: B** — show_details => true returns an OBJECT with usage.prompt_tokens, completion_tokens, total_tokens.

**Q3.** What is the correct syntax for AI_COMPLETE with a single image?

A) AI_COMPLETE(model, PROMPT(text, file))
B) AI_COMPLETE(model, predicate, TO_FILE(stage, path))
C) AI_COMPLETE(model, file, text)
D) AI_COMPLETE(model, text, {image: file})

> **Answer: B** — Single image: AI_COMPLETE(model, predicate_string, TO_FILE('@stage','file.jpg')). 3 positional args.

**Q4.** Which function replaces both EMBED_TEXT_768 and EMBED_TEXT_1024?

A) AI_SIMILARITY
B) AI_EMBED
C) AI_COMPLETE
D) AI_EXTRACT

> **Answer: B** — AI_EMBED(model, input) is the unified replacement. Model choice determines vector dimension.

**Q5.** AI_CLASSIFY returns an OBJECT. How do you extract the first label?

A) :label::STRING
B) :labels[0]::STRING
C) :category::STRING
D) :result::STRING

> **Answer: B** — AI_CLASSIFY returns {labels: [...]}. Use :labels[0]::STRING for single-label, :labels for multi-label array.

**Q6.** What does AI_SENTIMENT return?

A) FLOAT between -1 and 1
B) STRING ('positive'/'negative')
C) OBJECT with :categories array containing name+sentiment pairs
D) BOOLEAN

> **Answer: C** — AI_SENTIMENT returns {categories: [{name:'overall', sentiment:'positive'}, ...]}. Not a float like the old SENTIMENT function.

**Q7.** How do you get aspect-based sentiment for 'quality' and 'price'?

A) AI_SENTIMENT(text, 'quality, price')
B) AI_SENTIMENT(text, ['quality', 'price'])
C) AI_CLASSIFY(text, ['quality', 'price'])
D) AI_EXTRACT(text, ['quality', 'price'])

> **Answer: B** — Pass an array of category strings as the 2nd arg. Each gets its own sentiment in the response.

**Q8.** How many responseFormat types does AI_EXTRACT support?

A) 1
B) 3
C) 5
D) 7

> **Answer: C** — Simple object, array of strings, array of arrays, JSON schema (string type), JSON schema (array/table type).

**Q9.** What does AI_FILTER return?

A) STRING
B) OBJECT
C) BOOLEAN
D) ARRAY

> **Answer: C** — AI_FILTER returns TRUE or FALSE. Use in WHERE clause as a semantic filter.

**Q10.** How do you use AI_FILTER with two columns?

A) AI_FILTER(col1, col2)
B) AI_FILTER(PROMPT('template {0} {1}', col1, col2))
C) AI_FILTER(col1 || col2)
D) AI_FILTER(col1, 'filter', col2)

> **Answer: B** — PROMPT() creates a formatted input string with positional placeholders for multi-column filtering.

**Q11.** What is the correct syntax for AI_TRANSLATE with auto-detect?

A) AI_TRANSLATE(text, 'auto', 'en')
B) AI_TRANSLATE(text, NULL, 'en')
C) AI_TRANSLATE(text, '', 'en')
D) AI_TRANSLATE(text, 'en')

> **Answer: C** — Empty string '' for source language triggers automatic detection. NULL and 'auto' are not valid. The function requires three arguments: text, source_language, target_language.

**Q12.** AI_SUMMARIZE_AGG is what type of function?

A) Scalar
B) Aggregate
C) Table
D) Window

> **Answer: B** — It's an aggregate function like SUM or COUNT. Requires GROUP BY if other columns are selected.

**Q13.** What are the two arguments to AI_AGG?

A) model, prompt
B) column, model
C) text_expr, instruction_string
D) column, options_object

> **Answer: C** — AI_AGG(text_expression, instruction). Both must be strings. Use || to concatenate multiple columns.

**Q14.** What PII categories does AI_REDACT use by default?

A) Only NAME and EMAIL
B) Only PHONE_NUMBER
C) All supported categories
D) None — you must specify

> **Answer: C** — With no categories arg, AI_REDACT detects and redacts all supported PII types.

**Q15.** How do you use AI_REDACT in detect mode?

A) AI_REDACT(text, 'detect')
B) AI_REDACT(text, [], FALSE, 'detect')
C) AI_REDACT(input => text, mode => 'detect')
D) AI_REDACT(text, mode='detect')

> **Answer: C** — Named argument syntax: AI_REDACT(input => text, mode => 'detect'). Positional with empty array returns null.

**Q16.** What does AI_PARSE_DOCUMENT's LAYOUT mode preserve that OCR doesn't?

A) Color information
B) Document structure (headings, tables, sections)
C) Image resolution
D) Font styles

> **Answer: B** — LAYOUT preserves structural elements. OCR extracts raw text only.

**Q17.** What encryption type must a stage have for TO_FILE() to work?

A) SNOWFLAKE_FULL
B) AWS_CSE
C) SNOWFLAKE_SSE
D) No encryption required

> **Answer: C** — Stages must use SNOWFLAKE_SSE encryption. Client-side encrypted stages don't work with AI functions.

**Q18.** What does CREATE CORTEX SEARCH SERVICE require?

A) An external search engine
B) A pre-built embedding table
C) An ON column, WAREHOUSE, and TARGET_LAG
D) A fine-tuned model

> **Answer: C** — Minimum: ON column (text to search), WAREHOUSE (for indexing), TARGET_LAG (refresh interval).

**Q19.** What is a semantic model in Cortex Analyst?

A) A trained ML model
B) A YAML file defining tables, dimensions, metrics, and verified queries
C) An embedding model
D) A vector index

> **Answer: B** — Semantic models give Cortex Analyst business context to generate accurate SQL from natural language.

**Q20.** Which function has NO AI_ prefixed equivalent yet?

A) COMPLETE
B) SUMMARIZE
C) EXTRACT_ANSWER
D) TRANSLATE

> **Answer: C** — EXTRACT_ANSWER, TRY_COMPLETE, SPLIT_TEXT_RECURSIVE_CHARACTER, and FINETUNE have no AI_ equivalent.

**Q21.** What replaced SNOWFLAKE.CORTEX.SUMMARIZE?

A) AI_COMPLETE
B) AI_SUMMARIZE_AGG
C) AI_AGG
D) AI_EXTRACT

> **Answer: B** — AI_SUMMARIZE_AGG replaces SUMMARIZE. It's an aggregate function that handles datasets larger than the context window.

**Q22.** In a Dynamic Table, should you use AI_SUMMARIZE_AGG(body) per row?

A) Yes, it works per-row
B) No, use AI_COMPLETE for per-row summarization — AI_SUMMARIZE_AGG is an aggregate
C) Yes, with a GROUP BY on every column
D) No, use AI_EXTRACT instead

> **Answer: B** — AI_SUMMARIZE_AGG is aggregate. For per-row summarization in DTs, use AI_COMPLETE with a summarize prompt.

**Q23.** What is the response_format argument in AI_COMPLETE used for?

A) Setting the output language
B) Forcing structured JSON output with a schema
C) Choosing the model
D) Enabling guardrails

> **Answer: B** — response_format => {'type':'json','schema':{...}} forces the model to return structured JSON matching the schema.

**Q24.** Where does response_format go in AI_COMPLETE's named argument syntax?

A) Inside model_parameters
B) As a separate named argument: response_format => ...
C) In the prompt string
D) It's not supported

> **Answer: B** — response_format is its own top-level named argument, NOT nested inside model_parameters.

**Q25.** What does PROMPT() function do?

A) Generates a prompt from a template
B) Creates a prompt object with positional placeholders ({0}, {1}) for TO_FILE() references
C) Validates a prompt
D) Counts prompt tokens

> **Answer: B** — PROMPT('template {0} {1}', file1, file2) creates a prompt object mixing text and file references.

**Q26.** How does AI_SIMILARITY differ from manual cosine similarity?

A) It's faster
B) It embeds both inputs and computes similarity in one call — no separate AI_EMBED needed
C) It uses a different metric
D) It only works with images

> **Answer: B** — AI_SIMILARITY handles embedding + comparison internally. Manual approach requires AI_EMBED + VECTOR_COSINE_SIMILARITY.

**Q27.** What does AI_CLASSIFY's output_mode: 'multi' do?

A) Returns multiple models' results
B) Allows multiple labels to be assigned to a single input
C) Processes multiple rows at once
D) Returns probabilities

> **Answer: B** — Multi-label mode returns all matching categories in the labels array, not just the top one.

**Q28.** What is the EXTRACT_ANSWER function used for?

A) Extracting named entities
B) Pulling the answer to a question from source text (extractive QA)
C) Extracting images from PDFs
D) Filtering rows

> **Answer: B** — EXTRACT_ANSWER performs extractive QA — it finds and returns the answer span from the source text.

**Q29.** What does AI_EMBED's model choice determine?

A) The output language
B) The vector dimension (768 or 1024)
C) The cost per token
D) The maximum input length

> **Answer: B** — e5-base-v2 → 768-dim. snowflake-arctic-embed-l-v2.0 → 1024-dim. Model determines the vector size.

**Q30.** How do you chain translate and summarize across languages?

A) AI_SUMMARIZE_AGG(AI_TRANSLATE(text, src, 'en'))
B) AI_TRANSLATE(AI_SUMMARIZE_AGG(text), 'en')
C) AI_COMPLETE('translate and summarize: ' || text)
D) AI_AGG(text, 'translate and summarize')

> **Answer: A** — Translate first, then summarize the translated text. AI_SUMMARIZE_AGG wraps AI_TRANSLATE as the inner expression.

**Q31.** What is the DIRECTORY setting on a stage used for?

A) Organizing files into folders
B) Enabling DIRECTORY(@stage) queries that list staged files by RELATIVE_PATH
C) Automatic file compression
D) Setting access permissions

> **Answer: B** — DIRECTORY = (ENABLE = TRUE) allows FROM DIRECTORY(@stage) which returns metadata including RELATIVE_PATH.

**Q32.** What Cortex Agent tool type connects to Cortex Search?

A) cortex_analyst_tool
B) cortex_search_tool
C) sql_tool
D) function_tool

> **Answer: B** — cortex_search_tool connects a search service. cortex_analyst_tool connects a semantic model.

**Q33.** What is the difference between AI_EXTRACT and EXTRACT_ANSWER?

A) They are the same
B) AI_EXTRACT pulls structured fields; EXTRACT_ANSWER performs question-answering from text
C) EXTRACT_ANSWER is newer
D) AI_EXTRACT only works on images

> **Answer: B** — AI_EXTRACT extracts named fields (entity extraction). EXTRACT_ANSWER finds the answer to a question in text (extractive QA).

**Q34.** What does AI_COUNT_TOKENS('ai_classify', ...) count?

A) Output tokens only
B) Input tokens including the categories and config overhead
C) The number of classifications
D) The number of labels

> **Answer: B** — AI_COUNT_TOKENS includes all input: text + categories + config. This is the billable token count.

**Q35.** What happens when AI_COMPLETE cannot process an input row?

A) The entire query fails
B) The row returns NULL (default) or an error OBJECT if return_error_details is TRUE
C) The row is skipped
D) An exception is raised

> **Answer: B** — By default, failed rows return NULL without failing the query. return_error_details => TRUE returns {value, error}.
