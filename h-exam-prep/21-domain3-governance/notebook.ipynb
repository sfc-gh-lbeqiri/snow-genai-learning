## Practice Questions

**Q1.** Which database role is required to call Cortex AI functions?

A) SYSADMIN
B) SNOWFLAKE.CORTEX_USER
C) ACCOUNTADMIN
D) SECURITYADMIN

> **Answer: B** — SNOWFLAKE.CORTEX_USER database role must be granted to any role that calls AI functions.

**Q2.** What values can CORTEX_MODELS_ALLOWLIST accept?

A) Only model names
B) 'All', 'None', or a comma-separated list of model names
C) TRUE or FALSE
D) A JSON array

> **Answer: B** — Three formats: 'All' (all models), 'None' (no models), or 'mistral-large2,llama3.1-70b' (specific list).

**Q3.** Who can set the CORTEX_MODELS_ALLOWLIST parameter?

A) Any role with CORTEX_USER
B) SYSADMIN
C) ACCOUNTADMIN only
D) SECURITYADMIN

> **Answer: C** — CORTEX_MODELS_ALLOWLIST is an account-level parameter. Only ACCOUNTADMIN can alter it.

**Q4.** What does CORTEX_ENABLED_CROSS_REGION allow?

A) Data replication across regions
B) Running models in other Snowflake regions when not available locally
C) Multi-region failover
D) Cross-database queries

> **Answer: B** — It enables cross-region inference so models unavailable in your region can run in another Snowflake region.

**Q5.** How do you enable per-model RBAC?

A) GRANT USAGE ON MODEL
B) CALL SNOWFLAKE.MODELS.CORTEX_BASE_MODELS_REFRESH() then grant application roles
C) Set CORTEX_MODELS_ALLOWLIST
D) ALTER MODEL SET RBAC = TRUE

> **Answer: B** — Refresh populates SNOWFLAKE.MODELS with model objects and creates application roles. Then GRANT the roles.

**Q6.** How are Cortex AI functions billed?

A) Per query
B) Per token (input + output) at model-specific credit rates
C) Flat monthly fee
D) Per warehouse-hour only

> **Answer: B** — Billing is per token. Different models have different credit rates. Specialized functions are cheaper than general AI_COMPLETE.

**Q7.** Which function pre-estimates token cost without incurring token billing?

A) AI_COMPLETE with show_details
B) AI_COUNT_TOKENS
C) AI_FILTER
D) SYSTEM$ESTIMATE_COST

> **Answer: B** — AI_COUNT_TOKENS only uses compute credits. It does not bill for tokens.

**Q8.** Which ACCOUNT_USAGE view shows per-function Cortex AI credit usage?

A) QUERY_HISTORY
B) METERING_DAILY_HISTORY
C) CORTEX_FUNCTIONS_USAGE_HISTORY
D) WAREHOUSE_METERING_HISTORY

> **Answer: C** — CORTEX_FUNCTIONS_USAGE_HISTORY breaks down credits by function name (AI_COMPLETE, AI_CLASSIFY, etc).

**Q9.** Is AI_SENTIMENT cheaper than AI_COMPLETE for sentiment analysis?

A) No, they cost the same
B) Yes, specialized functions are generally cheaper per token
C) AI_SENTIMENT is more expensive
D) There is no cost difference

> **Answer: B** — Specialized AI functions (SENTIMENT, CLASSIFY) use optimized models and are cheaper than general-purpose AI_COMPLETE.

**Q10.** What does Cortex Guard use under the hood?

A) GPT-4
B) Llama Guard 3
C) Claude Haiku
D) Snowflake Arctic

> **Answer: B** — Cortex Guard is built on Meta's Llama Guard 3 model. It evaluates model output before returning it.

**Q11.** How do you enable Cortex Guard in AI_COMPLETE?

A) AI_COMPLETE(model, prompt, {guardrails: true})
B) ALTER SESSION SET GUARDRAILS = TRUE
C) GRANT GUARD_USER TO ROLE
D) AI_GUARD(AI_COMPLETE(...))

> **Answer: A** — Pass guardrails: true in the model_parameters object (3rd arg) or via named parameter.

**Q12.** What additional token type appears when guardrails are enabled?

A) filter_tokens
B) guardrail_tokens
C) safety_tokens
D) review_tokens

> **Answer: B** — The usage metadata includes guardrail_tokens when Cortex Guard is active. This is an additional cost.

**Q13.** What is the best defense against prompt injection in Snowflake?

A) Input validation only
B) Cortex Guard (guardrails: true) + careful prompt design
C) Encrypting the prompt
D) Using a smaller model

> **Answer: B** — Combine guardrails for runtime protection with well-structured prompts that resist override attempts.

**Q14.** How do you detect hallucination in AI outputs?

A) Check token count
B) Use grounded prompts ('answer only from context') + RAG + guardrails
C) Compare output length
D) Run AI_FILTER on the output

> **Answer: B** — Grounded prompts + RAG context + Cortex Guard provide multi-layer hallucination defense.

**Q15.** What does AI_REDACT do with no categories argument?

A) Returns the text unchanged
B) Redacts only NAME
C) Detects and redacts all supported PII categories
D) Throws an error

> **Answer: C** — Default behavior: all PII types are detected and replaced with placeholders like [NAME], [EMAIL], [PHONE_NUMBER].

**Q16.** What are the correct PII category names for AI_REDACT?

A) FIRST_NAME, EMAIL_ADDR
B) NAME, EMAIL, PHONE_NUMBER, ADDRESS
C) person, email, phone
D) PII_NAME, PII_EMAIL

> **Answer: B** — Uppercase category names: NAME, EMAIL, PHONE_NUMBER, ADDRESS, etc. Case-sensitive.

**Q17.** How do you use AI_REDACT's detect mode?

A) AI_REDACT(text, 'detect')
B) AI_REDACT(input => text, mode => 'detect')
C) AI_DETECT(text)
D) AI_REDACT(text, [], FALSE, 'detect')

> **Answer: B** — Named argument syntax only. Positional form with empty array returns null. mode => 'detect' returns PII spans.

**Q18.** What does AI_REDACT detect mode return?

A) Redacted text
B) BOOLEAN indicating PII presence
C) An OBJECT with a spans array listing each PII occurrence (category, start, end, text)
D) A count of PII items

> **Answer: C** — Detect mode returns {spans: [{category, start, end, text}, ...]} without modifying the original text.

**Q19.** Which ACCOUNT_USAGE view tracks Cortex AI query execution?

A) CORTEX_FUNCTIONS_USAGE_HISTORY
B) LOGIN_HISTORY
C) QUERY_HISTORY
D) TABLE_STORAGE_METRICS

> **Answer: C** — QUERY_HISTORY captures all queries including AI function calls. Filter by query_text patterns.

**Q20.** What is the LLM-as-Judge pattern?

A) Using an LLM to judge other users
B) Using AI_COMPLETE to automatically evaluate AI-generated answers against a context/rubric
C) A Snowflake built-in evaluation function
D) A model fine-tuning technique

> **Answer: B** — Call AI_COMPLETE with an evaluation prompt + context + question + answer. Score for accuracy and completeness.

**Q21.** What are Event Tables used for in Cortex AI observability?

A) Storing AI function results
B) Capturing telemetry, logs, and traces from UDFs and procedures
C) Tracking model versions
D) Billing breakdowns

> **Answer: B** — Event tables capture log/trace events. Set log_level and trace_level on functions/procedures to control verbosity.

**Q22.** What is the best practice for the DS_ROLE and database creation?

A) Grant CREATE DATABASE ON ACCOUNT to DS_ROLE
B) SYSADMIN creates the database and grants OWNERSHIP to DS_ROLE
C) DS_ROLE creates everything as ACCOUNTADMIN
D) Use SECURITYADMIN for all object creation

> **Answer: B** — Least privilege: SYSADMIN creates the database, grants OWNERSHIP to the custom role. No account-level grants needed.

**Q23.** What role hierarchy best practice does Snowflake recommend?

A) All roles inherit from ACCOUNTADMIN
B) Custom roles are granted to SYSADMIN to maintain the standard hierarchy
C) Each user gets ACCOUNTADMIN
D) No hierarchy needed

> **Answer: B** — Custom roles should be granted to SYSADMIN. ACCOUNTADMIN stays at the top, used sparingly.

**Q24.** How do you restrict a role to only use specific Cortex models?

A) CORTEX_MODELS_ALLOWLIST per role
B) Create model objects with CORTEX_BASE_MODELS_REFRESH, then grant model-specific application roles
C) Set session parameters
D) Use row access policies

> **Answer: B** — RBAC via SNOWFLAKE.MODELS: refresh model objects, then GRANT APPLICATION ROLE SNOWFLAKE."CORTEX-MODEL-ROLE-<MODEL>" TO ROLE.

**Q25.** What is the recommended order for cost optimization with AI functions?

A) Fine-tune first, then prompt
B) Use the largest model for accuracy
C) Start with specialized functions (AI_SENTIMENT, AI_CLASSIFY), use AI_COMPLETE only when needed, pre-estimate with AI_COUNT_TOKENS
D) Always use guardrails for cost control

> **Answer: C** — Specialized functions are cheaper. Estimate with AI_COUNT_TOKENS. Use AI_COMPLETE as fallback for tasks without a dedicated function.
